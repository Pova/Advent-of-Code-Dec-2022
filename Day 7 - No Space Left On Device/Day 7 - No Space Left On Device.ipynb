# Day 7 - No Space Left On Device

## Part 1

```python
with open('input_day7.txt') as f_in:
    data = f_in.read()
    
class Directory:
    def __init__(self,name,parent=None):
        self.name = name
        self.parent = parent
        self.files = []
        #key = name, value = node
        self.subdirs = {}
        self.size = 0
        
    def addFiles(self,file_name,file_size):
        print(f'add file: {file_name} to {self.name} (size {file_size})')
        self.files.append((file_name,file_size))
        self.updateDirSize(file_size)
        
    def addSubDirectories(self,dir_name):
        print(f'add subdirectory {dir_name} to node {self.name}')
        self.subdirs[dir_name] = Directory(dir_name,self)
        
    def updateDirSize(self,file_size):
        #print(f'updating the size of {self.name} by {file_size}')
        self.size += int(file_size)
        #print(f'Updated Size({self.name}) = {self.size}')
        if self.parent != None:
            self.parent.updateDirSize(file_size)
        
    
    
lines = data.split('\n')[:-1]

test = [
'$ cd /',
'$ ls',
'dir a',
'14848514 b.txt',
'8504156 c.dat',
'dir d',
'$ cd a',
'$ ls',
'dir e',
'29116 f',
'2557 g',
'62596 h.lst',
'$ cd e',
'$ ls',
'584 i',
'$ cd ..',
'$ cd ..',
'$ cd d',
'$ ls',
'4060174 j',
'8033020 d.log',
'5626152 d.ext',
'7214296 k'
]

#lines = test
            
def populateDirectory(line, cwd):
    #print(f'populateDirectory called on node {cwd.name}')
    if line[:3] == 'dir':
        cwd.addSubDirectories(line[4:])
    else:
        line_data = line.split()
        file_name = line_data[1]
        file_size = line_data[0]
        cwd.addFiles(file_name,file_size)

    
def changeDirectory(line,cwd):
    if line[0] == '$':
        #command
        if line[2:4] == 'cd':
            #print('changing directory...')
            #change directory
            if line == '$ cd ..':
                print(f'going up a level: old cwd = {cwd.name}, new cwd = {cwd.parent.name}')
                cwd = cwd.parent
            else:
                print(f'going down a level: old cwd = {cwd.name}, new cwd = {cwd.subdirs[line[5:]].name}')
                cwd = cwd.subdirs[line[5:]]
        else:
            print('shouldnt have any $ ls commands here!')
    else:
        print('shouldnt have any file listings here!')
    return cwd

    
list_idx = 1
dir_tree = Directory('/',None) #initialize node tree
cwd = dir_tree #initilize cwd

while list_idx < len(lines):
    #print('------------------------------')
    #print(f'beginning of while loop, idx = {list_idx}:')
    #print()
    
    if lines[list_idx][:4] == '$ ls':
        #print('within $ls command...')
        list_idx += 1
        #print(f'starting idx = {list_idx}')
        while list_idx<len(lines) and lines[list_idx][0] != '$':
            populateDirectory(lines[list_idx], cwd)
            list_idx += 1
        #print(f'ending idx = {list_idx}')
        if list_idx == len(lines):
            print('reached end of the list of instructions')
            break
        else:
            #print('testing continue statement...')
            continue 

    else:
        #print('change directory command')
        cwd = changeDirectory(lines[list_idx],cwd)
    
    list_idx += 1
    
    
```

## Part 2

```python
sum_of_sizes = 0

def printNodeSizes(tree):
    global sum_of_sizes
    if tree != None:
        if tree.size<100000:
            sum_of_sizes += tree.size
        print(f'Directory = {tree.name}, size = {tree.size}')
    for node in tree.subdirs.keys():
        printNodeSizes(tree.subdirs[node])

printNodeSizes(dir_tree)
print(f'final sum = {sum_of_sizes}')
```

## Part 2

```python
total_size = dir_tree.size
unused_space = 70000000-dir_tree.size
size_to_remove = 30000000 - unused_space
best_to_delete = dir_tree.size #initialized as whole tree

print(f'initial best = {best_to_delete}')

def findBestToDelete(tree):
    global size_to_remove
    global best_to_delete
    if tree != None:
        #print(f'Directory = {tree.name}, size = {tree.size}')
        if tree.size>size_to_remove and tree.size<best_to_delete:
            best_to_delete = tree.size
    for node in tree.subdirs.keys():
        findBestToDelete(tree.subdirs[node])

findBestToDelete(dir_tree)
print(f'final best = {best_to_delete}')
```
