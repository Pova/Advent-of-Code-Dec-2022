```python
import curses
import time
from curses import wrapper

with open('input_day9.txt') as f_in:
    data = f_in.read()
lines = data.split('\n')[:-1]

test = [
'R 4',
'U 4',
'L 3',
'D 1',
'R 4',
'D 1',
'L 5',
'R 2'
]


test_2 = [
'R 5',
'U 8',
'L 8',
'D 3',
'R 17',
'D 10',
'L 25',
'U 20'
]

#lines = test
#lines = test_2

#lines

#get bounds for the path:

min_i = 0 #x
max_i = 0 #x
min_j = 0 #y
max_j = 0 #y

i,j = 0,0
for line in lines:
    move = line.split()
    if move[0] == 'R':
        i += int(move[1])
        if i > max_i:
            max_i = i
    elif move[0] == 'D':
        j -= int(move[1])
        if j < min_j:
            min_j = j
    elif move[0] == 'L':
        i -= int(move[1])
        if i < min_i:
            min_i = i
    else:
        #move[0] == 'U'
        j += int(move[1])
        if j > max_j:
            max_j = j

max_width = max_i - min_i + 1
max_height = max_j - min_j + 1

#print(f'min_i = {min_i}, max_i = {max_i}, min_j = {min_j}, max_j = {max_j}')
#print(f'max_width = {max_width}, max_height = {max_height}')

move_list = []
for line in lines:
    move = line.split()[0]
    count = line.split()[1]
    for _ in range(int(count)):
        move_list.append(move)

#print(move_list)

grid = [['.' for _ in range(max_width)] for _ in range(max_height)]

#print(grid)

start = {'x':0,'y':0}
head = {'x':0,'y':0}
tail = {'x':0,'y':0}

head_position = [(head['x'],head['y'])]
tail_position = [(tail['x'],tail['y'])]
#tail_positions = {(tail['x'],tail['y'])}


#print(pos_dict.keys())

def checkDist(head,tail):
    dist_x = abs(head['x']-tail['x'])
    dist_y = abs(head['y']-tail['y'])
    if dist_x>1 or dist_y>1:
        moveTail(head,tail,dist_x,dist_y)



def moveHead(head,move):
    if move=='U':
        head['y'] += 1
    elif move=='R':
        head['x'] += 1
    elif move=='D':
        head['y'] -= 1
    elif move=='L':
        head['x'] -= 1
    else:
        pass
        #do nothing


def moveTail(head, tail,dist_x,dist_y):

    if dist_x == 0:
        #move in y direction
        if head['y'] > tail['y']:
            tail['y'] +=1
        else:
            tail['y'] -=1
    elif dist_y == 0:
        #move in x direction
        if head['x'] > tail['x']:
            tail['x'] +=1
        else:
            tail['x'] -=1
    elif dist_x > dist_y:
        if head['x'] > tail['x']:
            tail['x'] +=1
        else:
            tail['x'] -=1
        tail['y'] = head['y']
    elif dist_x < dist_y:
        if head['y'] > tail['y']:
            tail['y'] +=1
        else:
            tail['y'] -=1
        tail['x'] = head['x']
    elif dist_x == dist_y:
        if head['y'] > tail['y'] and head['x'] > tail['x']:
            tail['x'] += 1
            tail['y'] += 1
        elif head['y'] > tail['y'] and head['x'] < tail['x']:
            tail['x'] -= 1
            tail['y'] += 1
        elif head['y'] < tail['y'] and head['x'] > tail['x']:
            tail['x'] += 1
            tail['y'] -= 1
        elif head['y'] < tail['y'] and head['x'] < tail['x']:
            tail['x'] -= 1
            tail['y'] -= 1
        else:
            print('THIS SHOULDNT HAPPEN')
            print(head,tail)
    else:
        print('THIS SHOULDNT HAPPEN')
        print(head,tail)

for move in move_list:
    moveHead(head,move)
    head_position.append((head['x'],head['y']))
    checkDist(head,tail)
    tail_position.append((tail['x'],tail['y']))

# -------------------------

print(f'Total positions of tail piece = {len(set(tail_position))}')
```
