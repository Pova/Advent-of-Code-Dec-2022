```python
import copy

with open('input_day14.txt') as f_in:
    data = f_in.read()
lines = data.split('\n')[:-1]
lines

with open('test_day14.txt') as f_in:
    data = f_in.read()
test_lines = data.split('\n')[:-1]
#lines = test_lines

def flatten(l):
    return [item for sublist in l for item in sublist]

def draw_grid(grid_list):
    line_str = ''
    for line in grid_list:
        line_str += ''.join(line)
        line_str += '\n'
    print(line_str)

sand_source = [500,0]

#cleaning up lines
clean_lines = []
for line in lines:
    positions = line.split('->')
    positions = [tuple([int(digit) for digit in item.strip().split(',')]) for item in positions]
    clean_lines.append(positions)    

lines = clean_lines
lines_copy = copy.deepcopy(lines)


for line in lines:
    for i in range(len(line)-1):
        if line[i][0] == line[i+1][0]:
            #Same x position
            for idx in range(min(line[i][1],line[i+1][1])+1,max(line[i][1],line[i+1][1])):
                line.append((line[i][0],idx))
        if line[i][1] == line[i+1][1]:
            #Same y position
            for idx in range(min(line[i][0],line[i+1][0])+1,max(line[i][0],line[i+1][0])):
                line.append((idx,line[i][1]))

# print(f'lines_copy = {lines_copy}')
# print()
# print(f'lines = {lines}')
# print()

min_x = float('inf')
max_x = float('-inf')
min_y = float('inf')
max_y = float('-inf')
for line in lines:
    for pos_tuple in line:
        if pos_tuple[0]<min_x:
            min_x = pos_tuple[0]
        elif pos_tuple[0]>max_x:
            max_x = pos_tuple[0]

        if pos_tuple[1]<min_y:
            min_y = pos_tuple[1]
        elif pos_tuple[1]>max_y:
            max_y = pos_tuple[1]

print('------------------------')
print(f'min_x, max_x = {min_x}, {max_x}')
print(f'min_y, max_y = {min_y}, {max_y}')
print(f'source at: ({sand_source[0]},{sand_source[1]})')
print('------------------------\n')

line_str_list = [['.' for _ in range(min_x-2,max_x+3)] for _ in range(max_y+3)]

for row_idx in range(len(line_str_list)):
    for col_idx in range(len(line_str_list[0])):
        if (col_idx + min_x-2,row_idx) in flatten(lines):
            line_str_list[row_idx][col_idx] = '#'
line_str_list[0][500-min_x+2] = '+'
for row_idx in range(len(line_str_list)):
    line_str_list[row_idx][0] = '|'
    line_str_list[row_idx][-1] = '|'
for col_idx in range(1,len(line_str_list[0])-1):
    line_str_list[-1][col_idx] = '_'
# line_str_list[-1][0] = "\\"
# line_str_list[-1][-1] = "/"

#draw_grid(line_str_list)

# While (some condition is satisfied...): 
#     - Generate a sand unit
#     - Simulate the drop
#     - If hit an obstacle... figure out where to move
#     - if it hits any kill planes BREAK WHILE LOOP
#     - else... it settles

def newPositionCalculation(initial_position,grid_list):
    global hit_kill_plane
    pos_x_idx = initial_position[0] - min_x + 2
    pos_y_idx = initial_position[1]
    
    #1 - try down
    if (grid_list[pos_y_idx+1][pos_x_idx] == '.'):
        #print('Directly below blocked')
        return [initial_position[0],initial_position[1]+1]
    elif (grid_list[pos_y_idx+1][pos_x_idx] == '|' or grid_list[pos_y_idx+1][pos_x_idx] =='_'):
        grid_list[pos_y_idx+1][pos_x_idx] = 'X'
        print('hit the kill plane!')
        hit_kill_plane = True
        return None
    
    #2 - try diagonal down-left
    if (grid_list[pos_y_idx+1][pos_x_idx-1] == '.'):
        #print('Diagonal below left blocked')
        return [initial_position[0]-1,initial_position[1]+1]
    elif (grid_list[pos_y_idx+1][pos_x_idx-1] == '|' or grid_list[pos_y_idx+1][pos_x_idx-1] =='_'):
        grid_list[pos_y_idx+1][pos_x_idx-1] = 'X'
        print('hit the kill plane!')
        hit_kill_plane = True
        return None
    
    #3 - try diagonal down-right
    if (grid_list[pos_y_idx+1][pos_x_idx+1] == '.'):
        #print('Diagonal below right blocked')
        return [initial_position[0]+1,initial_position[1]+1]
    elif (grid_list[pos_y_idx+1][pos_x_idx+1] == '|' or grid_list[pos_y_idx+1][pos_x_idx+1] =='_'):
        grid_list[pos_y_idx+1][pos_x_idx+1] = 'X'
        print('hit the kill plane!')
        hit_kill_plane = True
        return None
    
    return initial_position
    
hit_kill_plane = False    
sand_counter = 0
while hit_kill_plane != True:
    sand_unit_loc = sand_source.copy()
    move_counter = 0
    while sand_unit_loc != None:
        new_sand_unit_loc = newPositionCalculation(sand_unit_loc,line_str_list)
        if new_sand_unit_loc == sand_unit_loc:
            #print(f'Sand unit blocked @ {new_sand_unit_loc}')
            line_str_list[new_sand_unit_loc[1]][new_sand_unit_loc[0]-min_x+2] = 'o'
            break
        else:
            sand_unit_loc = new_sand_unit_loc
        #print(f'After {move_counter} iterations, position = {sand_unit_loc}')
        move_counter += 1
    if sand_unit_loc == None:
        #print('sand_unit_loc == None')
        pass
    #print()
    #draw_grid(line_str_list)
    sand_counter += 1
    #print(f'sand_counter = {sand_counter}')
print('\n------------------------')    
print(f'Final sand counter = {sand_counter-1}')
print('------------------------')
```
