# Day 11 - Monkey in the Middle

## Part 1

```python
with open('input_day11.txt') as f_in:
    data = f_in.read()
    
lines = data.split('\n')

with open('test_day11.txt') as test:
    test_data = test.read()
    
test_lines = test_data.split('\n')

#lines = test_lines

class Monkey():
    def __init__(self):
        self.items = []
        self.operation = None
        self.test = None
        self.inspection_count = 0
        
    def addItem(self,item):
        self.items.append(item)
    
    def addItems(self,items):
        item_list = items.split(',')
        item_list = [int(_) for _ in item_list]
        self.items = item_list
        
    def addOperation(self,operation_line):
        calculation = operation_line.split('=')[1]
        def operation_func(old):
            return eval(calculation)
        self.operation = operation_func
        #self.operation = lambda old : eval(calculation)
    
    def addTest(self,test_line):
        modulo = int(test_line.split()[-1])
        def modulo_test(x):
            if x % modulo == 0:
                return True
            else:
                return False
        self.test = modulo_test
        
    def addTargets(self,if_true,if_false):
        self.targets = (if_true,if_false)
        
    def testItem(self):
        self.inspection_count += 1
        item_to_test = self.items.pop(0)
        new_item_value = self.operation(item_to_test)
        new_item_value = new_item_value//3
        if self.test(new_item_value):
            monkeys[self.targets[0]].addItem(new_item_value)
        else:
            monkeys[self.targets[1]].addItem(new_item_value)

monkeys = []

for i in range(0,len(lines),7):
    new_monkey = Monkey()
    
    constructor_lines = lines[i:i+6]
    
    # items
    items_line = constructor_lines[1].split(':')[1].strip()
    new_monkey.addItems(items_line)
    
    # operation
    operation_line = constructor_lines[2].split(':')[1].strip()
    new_monkey.addOperation(operation_line)
    
    #test
    test_line = constructor_lines[3].split(':')[1].strip()
    new_monkey.addTest(test_line)
    
    if_true = int(constructor_lines[4].split()[-1])
    if_false = int(constructor_lines[5].split()[-1])
    new_monkey.addTargets(if_true,if_false)
    
    monkeys.append(new_monkey)
    
for round_num in range(20):
    #print(f'-------- \n ROUND: {round_num + 1} \n')
    for name, monkey in enumerate(monkeys):
        #print(f'monkey number: {name} \n')
        while monkey.items != []:
            #print(monkey.items)
            monkey.testItem()
    #print('By the end of the round: \n')
    #for i,monkey in enumerate(monkeys):
        #print(f'monkey {i} has items: {monkey.items}')
        
inspection_counts = []
for monkey in monkeys:
    inspection_counts.append(monkey.inspection_count)
inspection_counts.sort(reverse=True)    
print(f'Level of monkey business after 20 rounds = {inspection_counts[0]*inspection_counts[1]}')
```
